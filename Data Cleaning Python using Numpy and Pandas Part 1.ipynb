# Data Science/Analysis - Data Cleaning using Python

```python
import numpy as np
import pandas as pd
```


Pandas provides a handy way of removing unwanted columns or rows from a DataFrame with the drop() function. Let’s look at a simple example where we drop a number of columns from a DataFrame.

First, let’s create a DataFrame out of the CSV file ‘BL-Flickr-Images-Book.csv’. In the examples below, we pass a relative path to pd.read_csv, meaning that all of the datasets are in a folder named Datasets in our current working directory:




```python
df = pd.read_csv('Datasets/BL-Flickr-Images-Book.csv')
```

Using head() method of a dataframe. It gives top five records of the dataset. 
# Data Cleaning Step: Column Dropping
Lets look at the first five entries and make a decision about which columns we need to drop.

```python
df.head()
```

Some of the columns do not provide much importance to our furhter data analysis. Lets make a list of the columns we need to drop.

```python
columns_drop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Contributors', 'Issuance type', 'Shelfmarks']
```

```python
print(columns_drop)
```

Using drop() method from pandas library to drop the selected columns of the dataset.

# Pandas Dataframe Drop method Syntax:
DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
- Parameters:
 - labels: object or dict or list to be dropped from the dataframe
 - axis:  int or axis name. Whether to drop labels from the index (0 / ‘index’) or columns (1 / ‘columns’).
 - inplace: bool, default False. If True, do operation inplace and return None.
 
This tells Pandas that we want the changes to be made directly in our object and that it should look for the values to be dropped in the columns of the object.

```python
df.drop(columns_drop, inplace=True, axis=1)
```

Lets see if the changes are made.

```python
df.head()
```

# Tips for Removing columns from the dataset
- Tip 1: Passing the columns= parameter in drop() method instead of passing the object label which need to be dropped
   - df.drop(columns=columns_drop, inplace=True)
- Tip 2: If you know in advance which columns you’d like to retain, you can directly use "usecols" argument of "pd.read_csv()" method

# Data Cleaning Step : Setting up key Index for the Dataframe

It is important to have a primary key column in your dataset here which is 'identifier' field. Let check if identifier has all unique values.

```python
df['Identifier'].is_unique
```

```python
df = df.set_index('Identifier')
```

Unlike primary keys in SQL, a Pandas Index doesn’t make any guarantee of being unique, although many indexing and merging operations will notice a speedup in runtime if it is.
